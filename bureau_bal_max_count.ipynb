```python
import pandas as pd
import numpy as np
```

```python
df_balance = pd.read_csv('data/bureau_balance.csv')
```

```python
df_balance
```

```python
df_value_counts= df_balance.groupby(['SK_ID_BUREAU','STATUS'],sort=True)['STATUS'].count().to_frame()

```

```python
df_value_counts.rename(columns={'STATUS':'COUNT'},inplace=True)
```

```python
df_value_counts.reset_index(inplace=True)
```

```python
df_value_counts.columns
```

```python
df_max_status_count = df_value_counts.sort_values('COUNT',ascending=False).drop_duplicates(['SK_ID_BUREAU'])
```

```python
df_max_status_count.sort_values('SK_ID_BUREAU',ascending=True)
```

```python
df_max_status_count.drop(['COUNT'],axis=1,inplace=True)
```

```python
df_max_status_count.to_csv('bureau_max_status_count.csv',index=False)
```
